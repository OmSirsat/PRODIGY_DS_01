```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/india_population.csv")
df
```

```python
plt.figure(figsize=(12, 8))

# Line plot for Population over the years
plt.subplot(2, 2, 1)
plt.plot(df['Year'], df['Population'], marker='o')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population over the Years')
```

```python
# Bar plot for Yearly % Change
plt.subplot(2, 2, 2)
plt.bar(df['Year'], df['Yearly % Change'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Yearly % Change')
plt.title('Yearly % Change')
```

```python
# Bar plot for Median Age
plt.subplot(2, 2, 3)
plt.bar(df['Year'], df['Median Age'], color='green')
plt.xlabel('Year')
plt.ylabel('Median Age')
plt.title('Median Age')
```

```python
# Bar plot for Fertility Rate
plt.subplot(2, 2, 4)
plt.bar(df['Year'], df['Fertility Rate'], color='orange')
plt.xlabel('Year')
plt.ylabel('Fertility Rate')
plt.title('Fertility Rate')
```

```python
plt.tight_layout()
plt.show()
```
